```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

a = 0.114
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink','blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'green', 'red', 'purple', 'brown', 'pink','blue']

#flow_time = [0, 0.10, 0.41, 0.92, 1.64, 2.56, 3.68, 5.01, 6.54, 8.28, 10.22]
flow_time = [0, 8.28, 10.22]

def list_observ(observable, ft):
    observable_cfg = []
    for cfgb in range(1776, 2052+1, 4):
        file_path = f"/pscratch/sd/h/hari_8/nEDM_project_LANL/save_GF_h5/save_cfg_b/save_EQ_WilsonFlow_cfg_b_{cfgb}_.h5"
        h5_file = h5py.File(file_path, "r")
        path_to_observable_cfgb = f"/b_cfg_{cfgb}/{observable}/flow_time_{ft}_fm"
        observable_cfgb = h5_file[path_to_observable_cfgb][()] 
        observable_cfg.append(observable_cfgb)
    #for cfga in range(1336, 1560+1, 4): #1612
    #    path_to_observable_cfga = f"/a_cfg_{cfga}/{observable}/flow_time_{ft}_fm" #_fm
    #    observable_cfga = h5_file[path_to_observable_cfga][()]
    #    observable_cfg.append(observable_cfga)
    return observable_cfg


def plt_histogram(observable, flow_time_list):
    for idx, f_t in enumerate(flow_time_list):
        observable_list = list_observ(observable, f_t)
        plt.hist(observable_list, histtype='step', bins=10, density=True, color=colors[idx], label="$t=$ %.4f" % (f_t))
    plt.legend()
    plt.ylabel("Probability")
    plt.xlabel("$E$")
    plt.title("No. of cfg = 70;  a = 0.114 fm;  $m_{\pi} = 317$ MeV")
    plt.suptitle("Distribution of $E$ - %s"%(observable))
    plt.savefig("Distribution_of_%s.pdf"%(observable), dpi=50000)
    plt.show()
    return

def plt_histogram_minu_nearestinteger(observable, flow_time_list):
    for idx, f_t in enumerate(flow_time_list):
        observable_list = list_observ(observable, f_t)
        nearestinteger_diff_observable_list = [value - round(value) for value in observable_list]
        plt.hist(nearestinteger_diff_observable_list, histtype='step', bins=10, density=True, color=colors[idx],  label="$t=$ %.4f" % (f_t))
    plt.legend()
    plt.ylabel("Probability")
    plt.xlabel("$E-$NearestInteger$(E)$")
    plt.title("No. of cfg = 70;  a = 0.114 fm;  $m_{\pi} = 317$ MeV")
    plt.suptitle("Distribution of $E-E_{int}$ - %s"%(observable))
    plt.savefig("Distribution_of_NearestInteger_%s.pdf"%(observable), dpi=50000)
    plt.show()
    return

def plt_histogramQ(observable, flow_time_list):
    for idx, f_t in enumerate(flow_time_list):
        observable_list = list_observ(observable, f_t)
        plt.hist(observable_list, histtype='step', bins=20, density=True, color=colors[idx], label="$t=$ %.4f" % (f_t))
    plt.legend()
    plt.ylabel("Probability")
    plt.xlabel("$Q$")
    plt.title("No. of cfg = 70;  a = 0.114 fm;  $m_{\pi} = 317$ MeV")
    plt.suptitle("Distribution of $Q$ - %s"%(observable))
    plt.savefig("Distribution_of_%s.pdf"%(observable), dpi=50000)
    plt.show()
    return

def plt_histogram_minu_nearestintegerQ(observable, flow_time_list):
    for idx, f_t in enumerate(flow_time_list):
        observable_list = list_observ(observable, f_t)
        nearestinteger_diff_observable_list = [value - round(value) for value in observable_list]
        plt.hist(nearestinteger_diff_observable_list, histtype='step', bins=20, density=True, color=colors[idx],  label="$t=$ %.4f" % (f_t))
    plt.legend()
    plt.ylabel("Probability")
    plt.xlabel("$Q-$NearestInteger$(Q)$")
    plt.title("No. of cfg = 70;  a = 0.114 fm;  $m_{\pi} = 317$ MeV")
    plt.suptitle("Distribution of $Q-Q_{int}$ - %s"%(observable))
    plt.savefig("Distribution_of_NearestInteger_int_%s.pdf"%(observable), dpi=50000)
    plt.show()
    return

plt_histogramQ("improvedQ", flow_time)
print("")
plt_histogram_minu_nearestintegerQ("improvedQ", flow_time)
print("")
plt_histogram("improvedE", flow_time)
print("")
plt_histogram_minu_nearestinteger("improvedE", flow_time)
```
